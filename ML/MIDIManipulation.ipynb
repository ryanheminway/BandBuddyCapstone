# Drum Timbre Manipulation via MIDI

We need the ability to take a MIDI file and swap out instruments to get different feels from the same drum track.



```python
import os
import sys
import pretty_midi
import pandas as pd
import IPython.display as ipd
from mido import MidiFile, MidiTrack

# Open original MIDI file
midi_path = os.path.join('..', 'groove', 'drummer1', 'session1','1_funk_80_beat_4-4.mid')
#input_midi = MidiFile(midi_path)
input_midi = pretty_midi.PrettyMIDI(midi_path)
midi_list = []

for instrument in input_midi.instruments:
    for note in instrument.notes:
        start = note.start
        end = note.end
        pitch = note.pitch
        velocity = note.velocity
        midi_list.append([start, end, pitch, velocity, "drum" if instrument.is_drum else "what"])
        
midi_list = sorted(midi_list, key=lambda x: (x[0], x[2]))

df = pd.DataFrame(midi_list, columns=['Start', 'End', 'Pitch', 'Velocity', 'Instrument'])
html = df.to_html(index=False)
ipd.HTML(html)
```
